```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Path to your CSV file
CSV_LOC = "/Users/nurupo/Desktop/dev/Project-Outlier-Music/dataset/music4all/stratified_songs_pop.csv"

# Load the dataset
df = pd.read_csv(CSV_LOC)

# Ensure 'release' is treated as an integer
df['release'] = df['release'].astype(int)

# Group by 'release' (year) and count the number of songs
songs_per_year = df.groupby('release').size().reset_index(name='number_of_songs')

# Plotting the distribution
plt.figure(figsize=(12, 6))
sns.barplot(x='release', y='number_of_songs', data=songs_per_year)

plt.title('Number of Songs per Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Songs', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

# If you have a lot of years, and the x labels overlap, you might want to sample them for better readability
# For example, to display every 5th year in the x-axis:
# plt.xticks(ticks=plt.xticks()[0], labels=[year if i % 5 == 0 else '' for i, year in enumerate(songs_per_year['release'])])

#plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Path to your CSV file
CSV_LOC = "/Users/nurupo/Desktop/dev/Project-Outlier-Music/dataset/music4all/stratified_songs_pop.csv"

# Load the dataset
df = pd.read_csv(CSV_LOC)

# Ensure 'release' is treated as an integer
df['release'] = df['release'].astype(int)

# Count the number of songs for each artist in each year
artist_songs_per_year = df.groupby(['release', 'artist']).size().reset_index(name='number_of_songs')

# Pivot the data to have years as the x-axis, artists as different colors, and song counts as the height of the bars
pivot_df = artist_songs_per_year.pivot(index='release', columns='artist', values='number_of_songs').fillna(0)

# Plotting the distribution
plt.figure(figsize=(12, 6))

# We use the 'bar' kind to create a bar plot
pivot_df.plot(kind='bar', stacked=True, figsize=(12, 6))

plt.title('Number of Songs per Artist Distributed by Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Songs', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='Artist', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
```
