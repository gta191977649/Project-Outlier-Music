```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Path to your CSV file
CSV_LOC = "/Users/nurupo/Desktop/dev/Project-Outlier-Music/dataset/music4all/stratified_songs_pop.csv"

# Load the dataset
df = pd.read_csv(CSV_LOC)

# Ensure 'release' is treated as an integer
df['release'] = df['release'].astype(int)

# Group by 'release' (year) and count the number of songs
songs_per_year = df.groupby('release').size().reset_index(name='number_of_songs')

# Plotting the distribution
plt.figure(figsize=(12, 6))
sns.barplot(x='release', y='number_of_songs', data=songs_per_year, palette='viridis')

plt.title('Number of Songs per Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Songs', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

# If you have a lot of years, and the x labels overlap, you might want to sample them for better readability
# For example, to display every 5th year in the x-axis:
# plt.xticks(ticks=plt.xticks()[0], labels=[year if i % 5 == 0 else '' for i, year in enumerate(songs_per_year['release'])])

#plt.show()
```
